```python
from scipy.io import netcdf
import numpy as np
import numpy.ma as ma
import pandas as pd
import netCDF4 as nc
```

```python
data = pd.read_excel("raw_data/maryland data.xlsx",'final data') # raw data
chesapeake_bay_area = 9.8*10**3*10**3 # 1000m^2
# densities
data.total_crabs = data.total_crabs*10**6 #millions of crabs
data.adult_fems = data.adult_fems*10**6 #millions of crabs
data.juveniles = data.juveniles*10**6 #millions of crabs
data['total_crabs_density'] = data.total_crabs/chesapeake_bay_area # crabs/1000m^2
data['adult_fems_density'] = data.adult_fems/chesapeake_bay_area # crabs/1000m^2
data['juveniles_density'] = data.juveniles/chesapeake_bay_area # crabs/1000m^2
#print(data)
```

```python
temperature = nc.Dataset("raw_data/tos.nwa.full.hcast.monthly.regrid.r20250715.199301-202312.nc")
tos = temperature.variables['tos'][:]
latitude = temperature.variables['lat'][:]
lat = np.round(latitude,2)
longitude = temperature.variables['lon'][:]
lon = np.round(longitude,2)
# chesapeake bay coordinates
chesapeake_lat = 38.21
chesapeake_lon = 283.74
x = np.argwhere(lon==chesapeake_lon)[0][0]
y = np.argwhere(lat==chesapeake_lat)[0][0]
T = np.reshape(tos[:,y,x],(12,-1))
```

```python
np.save('temperature.npy',T.data)
data[3:-1].to_csv('data.csv') # save from year 1993 to 2023 to match with temperature
```
